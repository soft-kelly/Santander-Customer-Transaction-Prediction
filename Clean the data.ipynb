### Check for Class Imbalance

```python
target = train['target']
sns.set_style('whitegrid')
sns.countplot(target)
```

```python
count = 0;
for i in target:
    if i == 1:
        count = count + 1;
print(count/len(target)*100,"% is 1,",100-count/len(target)*100,"% is 0. Thus the target in train is unbalanced.")
```

### Correlation Coefficient

```python
corr = train.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(train.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(train.columns)
ax.set_yticklabels(train.columns)
plt.show()
```

The correlation coefficient matrix told us the features are independent with each other.
